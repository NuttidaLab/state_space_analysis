```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# visualize a normal glm given the weights to the mean and variance
def plot_glm(weights, x, y, title):
    # plot the data
    plt.scatter(x, y, color='black', alpha=0.5)
    # plot the glm
    x = np.linspace(-3, 3, 100)
    y = weights[0] + weights[1] * x
    plt.plot(x, y, color='red')
    plt.title(title)
    plt.show()
```

```python
plot_glm(1, 1, 1, 'test')
```
