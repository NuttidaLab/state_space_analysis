```python
import os
from cmdstanpy import CmdStanModel
import arviz as az

import numpy as np
import pandas as pd
import xarray as xr

import matplotlib.pyplot as plt
import arviz as az
import seaborn as sns
```

```python
stan_file = os.path.join('mixture.stan')
model = CmdStanModel(stan_file=stan_file)
```

```python
S, T = 312, 120

rng = np.random.default_rng(seed=42)

x_tr = rng.normal(size=(T, 3))
x_em = rng.normal(size=(T))
obs = rng.normal(0, 1, size=(T))

data = {
    'T': T,

    'wmu': np.array([0, 4]),
    'intercepts': np.array([1, 2]),
    'kappa': np.array([1, 2]),

    'y': obs,

    'pi_init': np.array([0.5, 0.5]),

    'x_tr': x_tr,
    'x_em': x_em,
}
```

```python
fit = model.sample(data=data)
```

```python
az.plot_trace(fit)
```
