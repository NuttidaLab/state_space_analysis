```python
from scipy.io import loadmat
from scipy import stats
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Set in stone
data_path = '../axej/'
n_subjects = 13
n_sessions = 4
n_runs = 6
n_trials = 120
n_ts = 500

experiment_orientations = [159, 123, 87, 51, 15]
subjects = ["01", "02", "03", "04", "05", "06", "07", "08" ,"09", "10", "11", "12", "14"]

def get_calib(subj, sess):
    mat_contents = loadmat(data_path + f'AxeJEEG_Subj{subjects[subj]}_S{sess+1}_Cali1.mat.mat', struct_as_record=False, squeeze_me=True)
    return mat_contents["p"].__dict__

def get_run(subj, sess, run):
    mat_contents = loadmat(data_path + f'AxeJEEG_Subj{subjects[subj]}_S{sess+1}_Run{run+1}.mat.mat', struct_as_record=False, squeeze_me=True)
    return mat_contents["p"].__dict__
```

```python
# Get all calibration data
# subjexts x sessions
calib_jx = []
calib_jy = []
calib_sitmdir = []
calib_tgonset = []
for subj in tqdm(range(n_subjects)):
    for sess in range(n_sessions):
        calib = get_calib(subj, sess)

        calib_jx.append(calib["joyx"])
        calib_jy.append(calib["joyy"])
        calib_sitmdir.append(calib["stimDirREAL"])

        f_tgonset = calib["f_precuedur"] + calib["f_cuedur"]
        calib_tgonset.append(f_tgonset)

# Shape the calibration data
calib_jx = np.array(calib_jx, dtype=np.float64).reshape(n_subjects, n_sessions, n_trials, n_ts)
calib_jy = np.array(calib_jy, dtype=np.float64).reshape(n_subjects, n_sessions, n_trials, n_ts)
calib_sitmdir = np.array(calib_sitmdir).reshape(n_subjects, n_sessions, n_trials)
calib_tgonset = np.array(calib_tgonset).reshape(n_subjects, n_sessions, n_trials)

calib_jx.shape, calib_jy.shape, calib_sitmdir.shape, calib_tgonset.shape
```

```python
# preallocate empty array and assign slice by chrisaycock
def shift(arr, num, fill_value=np.nan):
    result = np.empty_like(arr)
    if num > 0:
        result[:num] = fill_value
        result[num:] = arr[:-num]
    elif num < 0:
        result[num:] = fill_value
        result[:num] = arr[-num:]
    else:
        result[:] = arr
    return result
```

```python
# shif by tgonset
shifted_jx = np.zeros_like(calib_jx)
shifted_jy = np.zeros_like(calib_jy)

shifted_jx[:] = np.nan
shifted_jy[:] = np.nan

for sub in range(n_subjects):
    for sess in range(n_sessions):
        for trial in range(n_trials):
            tgonset = calib_tgonset[sub, sess, trial]
            shifted_jx[sub, sess, trial, :] = shift(calib_jx[sub, sess, trial, :], -1*tgonset, fill_value=np.nan)
            shifted_jy[sub, sess, trial, :] = shift(calib_jy[sub, sess, trial, :], -1*tgonset, fill_value=np.nan)
```

```python
# Distance from center

calib_dist_from_cent = np.sqrt(calib_jx ** 2 + calib_jy ** 2)
calib_dist_from_cent[np.isnan(calib_dist_from_cent)] = 0

dist_from_cent = np.sqrt(shifted_jx ** 2 + shifted_jy ** 2)
dist_from_cent[np.isnan(dist_from_cent)] = 0
```

```python
# max distance indexs
max_idx = np.argmax(dist_from_cent, axis=-1)
```

```python
maxed_jx = np.zeros_like(shifted_jx)
maxed_jy = np.zeros_like(shifted_jy)

maxed_jx[:] = np.nan
maxed_jy[:] = np.nan

# Take everything till the max distance
for sub in range(n_subjects):
    for sess in range(n_sessions):
        for trial in range(n_trials):
            end_idx = max_idx[sub,sess,trial]
            maxed_jx[sub,sess,trial,:end_idx] = shifted_jx[sub,sess,trial,:end_idx]
            maxed_jy[sub,sess,trial,:end_idx] = shifted_jy[sub,sess,trial,:end_idx]
```

```python
def plot_runs(stimdir, x, y, sub, sess):
    plt.figure(figsize=(6, 4))
    for i in range(120): 
        if stimdir[sub,sess,i] == 159.0:
            color = 'r'
        elif stimdir[sub,sess,i] == 123.0:
            color = 'g'
        elif stimdir[sub,sess,i] == 87.0:
            color = 'b'
        elif stimdir[sub,sess,i] == 51.0:
            color = 'y'
        elif stimdir[sub,sess,i] == 15.0:
            color = 'k'

        plt.plot(x[sub,sess,i,:], y[sub,sess,i,:], '-', color=color, alpha=0.2)
    # outside the plot to the right
    plt.legend(['159 - red', '123 - green', '87 - blue', '51 - yellow', '15 - black'], loc='center left', bbox_to_anchor=(1, 0.5))
    plt.show()
```

```python
plot_runs(calib_sitmdir, maxed_jx, maxed_jy, 12, 3)
```

```python
# From the first instance of distance = 1, make everything after that nan
# This is to remove the overshoot

bound_jx = np.zeros_like(maxed_jx)
bound_jy = np.zeros_like(maxed_jy)

bound_jx[:] = np.nan
bound_jy[:] = np.nan

for sub in range(n_subjects):
    for sess in range(n_sessions):
        for trial in range(n_trials):
            clip_ts = 500

            for timestep in range(n_ts):
                if dist_from_cent[sub,sess,trial,timestep] >= 1:
                    # something bigger than one found
                    clip_ts = timestep
                    break

            bound_jx[sub,sess,trial,:clip_ts] = maxed_jx[sub,sess,trial,:clip_ts]
            bound_jy[sub,sess,trial,:clip_ts] = maxed_jy[sub,sess,trial,:clip_ts]
```

```python
# Calculate response angle
bound_resp = np.arctan2( bound_jy.flatten(), bound_jx.flatten() )
bound_resp = np.rad2deg(bound_resp)
bound_resp = (bound_resp + 360) % 360;
bound_resp = bound_resp.reshape(n_subjects, n_sessions, n_trials, n_ts)
```

```python
def plot_trial(resp,stimdir, jx, jy, sub, sess, trial):

    _, axs = plt.subplots(1, 2, figsize=(12, 4))

    axs[0].plot(resp[sub,sess,trial,:])
    axs[0].plot(stimdir[sub,sess,trial]*np.ones(500))
    axs[0].set(xlabel='TS', ylabel='Response Angle (deg)')

    # at each point in the plot, write the corresponding response angle
    axs[1].plot(jx[sub,sess,trial,:], jy[sub,sess,trial,:])
    for i in range(500):
        # skip if not a number
        if np.isnan(jx[sub,sess,trial,i]):continue
        axs[1].text(jx[sub,sess,trial,i], jy[sub,sess,trial,i], str(round(resp[sub,sess,trial,i], 2)), fontsize=8)
    axs[1].set(xlabel='JoyX', ylabel='JoyY')

    plt.show()
```

## looking at errors

```python
# get error angle over time

error_angle = np.zeros_like(bound_resp)
error_angle[:] = np.nan

for sub in range(n_subjects):
    for sess in range(n_sessions):
        for trial in range(n_trials):
            a = calib_sitmdir[sub,sess,trial]
            for ts in range(n_ts):
                if np.isnan(bound_resp[sub,sess,trial,ts]):continue
                b = bound_resp[sub,sess,trial,ts]
                error_angle[sub,sess,trial,ts] = min(abs(a-b), 360-abs(a-b))
```

```python
# find which complete run in erros has all null values

null_idx = []
for sub in range(n_subjects):
    for sess in range(n_sessions):
        for trial in range(n_trials):
            if np.all(np.isnan(error_angle[sub,sess,trial])):
                null_idx.append((sub, sess, trial))

null_idx
```

## Getting response angles

```python
def circdist(a, b):
    return min(abs(a-b), 360-abs(a-b))

def circmedian(angs):
    angs = np.array(angs)
    angs = angs[~np.isnan(angs)]
    pdists = angs[np.newaxis, :] - angs[:, np.newaxis]
    pdists = (pdists + 180) % (2 * 180) - 180
    pdists = np.abs(pdists).sum(1)
    return angs[np.argmin(pdists)]
```

```python
# Get the last non nan values as the response angle

last_resp_angle = np.zeros((n_subjects, n_sessions, n_trials))
last_resp_angle[:] = np.nan

for sub in range(n_subjects):
    for sess in range(n_sessions):
        for trial in range(n_trials):
            for ts in range(n_ts):
                if np.isnan(bound_resp[sub,sess,trial,ts]):
                    last_resp_angle[sub,sess,trial] = bound_resp[sub,sess,trial,ts - 1]
                    break
```

```python
# check if there are any nan values, these are to be ignored duing calculation
np.argwhere(np.isnan(last_resp_angle))
```

```python
# organize the response angles by stimdir
last_resp_angle.shape, calib_sitmdir.shape
```

```python
# 5 angles in total and each subject has 4 * 120/ 5 = 96 trials per angle
stim_resp = np.zeros((n_subjects, 5, 96))

for sub in range(n_subjects):
    # for each unique stimdir
    for i, unique_stim in enumerate(np.unique(calib_sitmdir)):
        # get the response angle for that stimdir by masking the calib_sitmdir with the unique stimdir
        stim_resp[sub, i] = last_resp_angle[sub][calib_sitmdir[sub]==unique_stim]

stim_resp.shape
```

```python
# now calculate the circular std and median for each subject and angle
```

```python
circ_median = np.zeros((n_subjects, 5))
circ_std = np.zeros((n_subjects, 5))

for sub in range(n_subjects):
    for i, unique_stim in enumerate(np.unique(calib_sitmdir)):
        circ_median[sub, i] = circmedian(stim_resp[sub, i])
        circ_std[sub, i] = np.rad2deg(stats.circstd(np.deg2rad(stim_resp[sub, i]), nan_policy='omit'))
```
